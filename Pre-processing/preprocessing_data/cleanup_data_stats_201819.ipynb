```python
################################################################################
# The following section is to identify players who got traded in the middle of a season and clean it up
# the players will have multiple rows
# pick the row that has TOT for team. 
################################################################################
```

```python
# Find Frequency of occurence of a players name
duplicate_name_df = player_stats_clean_df['player_id'].groupby(player_stats_clean_df['player_id']).count()
duplicate_name_df
```

```python
# Filter names if they are repeasted duplicate_name_df
for player_id in duplicate_name_df:
    print(player_id)
```

```python
duplicate_name_2_df.dtypes
```

```python
# Number of players who are traded (and have multiple rows)
duplicate_name_2_df.loc[(duplicate_name_2_df >1)].count()
```

```python
# Use the second method to create a column in 
player_stats_clean_df["player_frequency"]  = player_stats_clean_df.player_id.map(player_stats_clean_df.player_id.value_counts())
player_stats_clean_df
```

```python
# Filter Dataframe for Frequency = 1 
player_stats_clean_df.loc[(player_stats_clean_df['player_frequency'] == 1)]
```

```python
# Filter for Frequency !=1. When this  count is added up to frequency =1, total should match the total number of rows
player_stats_clean_df.loc[(player_stats_clean_df['player_frequency'] != 1)]
```

```python
# Filter Dataframe for Tm == "TOT and it should match with players who are traded (86)
player_stats_clean_df.loc[(player_stats_clean_df['Tm'] == "TOT")]
# count 86 matches. So, out filter is correct
```

```python
# Filter Dataframe for Frequency !=1 && Tm == "TOT"
#Total number of unique rows would be (frequency ==1) + (frequency !=1 & Tm = "TOT)")
#444 + 86 = 530
player_stats_filtered_df = player_stats_clean_df.loc[(player_stats_clean_df['player_frequency'] == 1) | (player_stats_clean_df['Tm'] == "TOT")]
player_stats_filtered_df
#Row count is 530 and it matches with what is expected
```

```python
##############################################################################
# The following section add additional columns needed 
# playing_season if value is 2018, then it maeans that the season is 2018-19
##############################################################################
player_stats_filtered_df['playing_season'] = "2018"
player_stats_filtered_df
```

```python
#############################################################################
# The following section creates a new data frame with just required columns 
# and rename them
############################################################################
```

```python
# List column names 
player_stats_filtered_df.columns
```

```python
# Create New DataFRame with only the columns that is needed
player_stats_cleaned_df = player_stats_filtered_df[['player_id','playing_season','Pos','Age','Tm','G','MP','FG','FG%','3P','3P%','eFG%','FT','FT%','TRB','AST','STL','BLK','TOV','PTS']]
player_stats_cleaned_df
```

```python
# Rename Column Names to match the one with the Database
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"Pos":"playing_position"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"Age":"playing_age"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"Tm":"team_name"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"G":"no_of_games_played"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"MP":"no_minutes_played"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"FG":"field_goals_made"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"FG%":"field_goals_pctg"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"3P":"three_pointers_made"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"3P%":"three_point_pctg"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"eFG%":"effective_field_goal_pctg"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"FT":"free_thows_made"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"FT%":"free_throws_pctg"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"TRB":"total_rebounds"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"AST":"no_of_assists"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"STL":"no_of_steals"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"BLK":"no_of_blocks"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"TOV":"no_of_turnovers"})
player_stats_cleaned_df = player_stats_cleaned_df.rename(columns={"PTS":"points"})
player_stats_cleaned_df
```
