```python
# Creating New Data Frame for 2016-17 with just the columns we need 
player_salary_201617_clean_df = player_salary_2016_17_df[['Player_ID','Season','Salary']]
player_salary_201617_clean_df
```

```python
# Change format of Season from 2016-17 to 2016
split_season_df = player_salary_201617_clean_df['Season'].str.split("-", n = 1, expand = True)
split_season_df
```

```python
# Add New Column for playing_season
player_salary_201617_clean_df["playing_season"] = split_season_df[0]
player_salary_201617_clean_df
```

```python
# Fill Salary Values with 0 if its null
player_salary_201617_clean_df[['Salary']] = player_salary_201617_clean_df[['Salary']].fillna(value=0)
player_salary_201617_clean_df
```

```python
# Change Column Names 
player_salary_201617_clean_df = player_salary_201617_clean_df.rename(columns={"Player_ID":"player_id"})
player_salary_201617_clean_df = player_salary_201617_clean_df.rename(columns={"Salary":"salary"})
player_salary_201617_clean_df
```

```python
# Create a Cleaned Data Frame that will be ouputted to a CSV file
player_salary_201617_cleaned_df = player_salary_201617_clean_df[['player_id','playing_season','salary']]
player_salary_201617_cleaned_df
```

```python
#########################
# Write to CSV File 2016
#########################
# writing Clean data to csv file
player_salary_201617_cleaned_df.to_csv(r'../../Resources/PreprocessedDataFiles/NBA Players Salary 201617_clean.csv', encoding='utf-8', header='true')
player_salary_201617_cleaned_df
```
