```python
# Read datasets from sample data (CSV File) - This is for temporary purpose till DB is ready and operational


# Read player stats and keeping them in dataframe
path= 'PLAYERS_STATS_F.csv'
players_stats_df=pd.read_csv(path)
players_stats_df=players_stats_df.set_index('player_id')

players_stats_df
```

```python
player_stats_avg_df= players_stats_df.groupby(['player_id']).mean()[['no_of_games_played','no_minutes_played','field_goals_pctg','three_point_pctg',''\
          'effective_field_goal_pctg','free_throws_pctg','total_rebounds','no_of_assists','no_of_steals','no_of_blocks',\
          'no_of_turnovers','points']]

player_stats_avg_df.head()

# Read player slary and keeping them in dataframe
path= 'PLAYERS_SALARIES_F.csv'
players_salaries_df=pd.read_csv(path)
players_salaries_df=players_salaries_df.set_index('player_id')
players_salaries_df = players_salaries_df.dropna()
#players_salaries_df.count()
players_salaries_avg_df= players_salaries_df.groupby(['player_id']).mean()['player_salary']

players_salaries_avg_df
```

```python
player_stats_with_salary = player_stats_avg_df.merge(players_salaries_avg_df, left_index=True,right_index=True)
player_stats_with_salary.head()
```

```python
def salary_increased_or_not_labeling(player_id):
    print ('coming inside function')
    player_first_year_salary = players_salaries_df[(players_salaries_df['player_id'] == player_id) & \
                                              (players_salaries_df['playing_season'] == '2016-2017')]['player_salary'].values
    
     
    #print(f'{player_id}\'s first year salary', player_first_year_salary)
    
    
    player_last_year_salary = players_salaries_df[(players_salaries_df['player_id'] == player_id) & \
                                              (players_salaries_df['playing_season'] == '2017-2018')]['player_salary'].values

    #print(f'{player_id}\'s last year salary', player_last_year_salary)

    
    if player_last_year_salary > player_first_year_salary:
        return 1
    else:
        return 0
```

```python
players_salaries_df = players_salaries_df.reset_index(['player_id'])
players_salaries_df
```

```python
player_first_year_salary = players_salaries_df[(players_salaries_df['player_id'] == 'acyqu01') & \
                                              (players_salaries_df['playing_season'] == '2016-2017')]['player_salary']
player_first_year_salary[1]
```

```python

import warnings
warnings.filterwarnings('ignore')


player_stats_with_salary_aggregated = player_stats_with_salary.reset_index(['player_id'])
player_stats_with_salary_aggregated['salary_increased_yorn'] = player_stats_with_salary_aggregated['player_id'].apply(salary_increased_or_not_labeling)
player_stats_with_salary_forML= player_stats_with_salary_aggregated.set_index(['player_id']).dropna()
player_stats_with_salary_forML


#player_stats_with_salary_aggregated['player_id'].apply(salary_increased_or_not_labeling)
```

```python
# Create our features
x_cols = [i for i in player_stats_with_salary_forML.columns if i not in ('salary_increased_yorn')]  
X = player_stats_with_salary_forML[x_cols]
X.describe()

y=player_stats_with_salary_forML['salary_increased_yorn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
y_test
```

```python

from imblearn.ensemble import BalancedRandomForestClassifier
clf = BalancedRandomForestClassifier(n_estimators=100,max_depth=None, random_state=1)
clf.fit(X_train, y_train)  

y_pred=clf.predict(X_test)
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# Calculated the balanced accuracy score
# YOUR CODE HERE


from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
# YOUR CODE HERE

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
