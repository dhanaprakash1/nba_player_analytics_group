```python
# Put player name into a list
player_id_name = dataframe_analyzed_df['Player'].tolist()

# Split the playid and name and list the rows which are not split into 2 parts  with \ as delimeter(either 1 or >2)
for name in player_id_name:
    print(name.split("\\"), len(name.split("\\")))
```

```python
# Splits the Name Column into two columns and add it to Dataframe
split_name_df = dataframe_analyzed_df['Player'].str.split("\\", n = 1, expand = True)
split_name_df
```
